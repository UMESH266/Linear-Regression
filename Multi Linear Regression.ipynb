#### Multiple Linear regressoin model fitting

##### Linear regression

```python
#1. Features extraction

# Predictor variables
x = start_up.drop('Profit', axis=1)

# Target variable 
y = start_up['Profit']
```

```python
# Converting State columns data type to numerical
print('Columnwise uniques values:\n', x.nunique())
x['State'] = x['State'].astype('category')
x['State'] = x['State'].cat.codes
```

```python
x.head()
```

```python
#2. splitting of dataset

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#3. Regressor model training
#a) Using sklearn library

regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
rmse_reg = (mean_squared_error(y_test, y_pred)) ** 0.5
r2 = r2_score(y_test, y_pred)
print('R squared value of Regressor model: ', round(r2, 4))
print('Root Mean Square Error of Regressor model: ', rmse_reg)
```

##### Ordinary Least Square

```python
#b) Using statsmodels library

# Adding constant to implement Ordinary Least Square model
X = sm.add_constant(x)
Y = y

# Split data into train and test
train_X, test_X, train_y, test_y = train_test_split(X, Y, train_size=0.8, random_state=0)

# Fit the regression model
mlr_model = sm.OLS(train_y, train_X).fit()
print(mlr_model.summary2())
```

Observations: 
1. Among Linear regression and Ordinary Least Squares (OLS) models, the OLS has higher R-square value of 0.95 compared to Linear regression model R-square value of 0.938. This shows that the OLS model explains the variation in dependent variable better than Linear regressoion.

2. The p-values for the t-test of all the features except R&D spend are above 0.05. This indicates that there is  not statistically significant relationship (at significance value of 0.05) between these features and Profit (dependent variable).  

3. The probability value of F-statistic of the model is 0.0016 (<0.05) which indicates that the overall model is statistically significant.

#### Observation of collinearity in the dataset

```python
# Calculating VIF
vif = pd.DataFrame()
vif["Variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print('Variance Inflation Factors - ')
vif
```

Observation: 

VIF is less than 5 for all independent variables. This shows that multicollinearity doesn't exists in the dataset.

#### OLS with Data transformations

##### Standard transformation

```python
# a) Standard transformation
# Data transformation
std_scaler = StandardScaler()
x_train_std = std_scaler.fit_transform(train_X)
x_train_std = sm.add_constant(x_train_std)
y_train_std = std_scaler.fit_transform(pd.DataFrame(train_y))

# Fit the regression model
std_model = sm.OLS(y_train_std, x_train_std).fit()
print(std_model.summary2())
```

##### MinMax Scaling

```python
# b) MinMax transformation
# Data transformation
minmax_scaler = MinMaxScaler()
x_train_mm = minmax_scaler.fit_transform(train_X)
x_train_mm = sm.add_constant(x_train_mm)
y_train_mm = minmax_scaler.fit_transform(pd.DataFrame(train_y))

# Fit the regression model
minmax_model = sm.OLS(y_train_mm, x_train_mm).fit()
print(minmax_model.summary2())
```

##### Squareroot transformation

```python
# c) Squareroot transformation

# Data transformation
x_train_sqrt = (train_X) ** 0.5
y_train_sqrt = (train_y) ** 0.5

# Fit the regression model
sqrt_model = sm.OLS(y_train_sqrt, x_train_sqrt).fit()
print(sqrt_model.summary2())
```

##### Outliers treated

```python
#d) Outlier analysis and transformation

# Model fitting
out_model = sm.OLS(Y, X).fit()

# Calculation of influence
influence = out_model.get_influence()
influence_list = influence.cooks_distance[0]
influence_df = pd.DataFrame(influence_list, columns=["Influence"])
influence_df.index = start_up.index
cooks_df = start_up.merge(influence_df, left_index=True, right_index=True)

# Setting the threshold and getting the outliers
original_length = len(start_up)
cooks_threshold = 4/original_length

cooks_outliers = cooks_df[cooks_df["Influence"] > cooks_threshold]

print('Details of outliers - ')
print("Removed:", len(cooks_outliers))
print(f"This is {len(cooks_outliers) / original_length * 100}% of our dataset")

# New dataset without outliers
new_data = cooks_df[cooks_df["Influence"] < cooks_threshold].reset_index()
new_data = new_data.drop('index', axis=1)
```

```python
# Improved data feature extraction
x_new = new_data.drop(['Profit', 'Influence'], axis=1)
y_new = new_data['Profit']

# Converting State columns data type to numerical
x_new['State'] = x_new['State'].astype('category')
x_new['State'] = x_new['State'].cat.codes

#Adding constant
x_new = sm.add_constant(x_new)
# Model fitting
new_data_model = sm.OLS(y_new, x_new).fit()
print(new_data_model.summary2())
```

```python
print('Summary of R^2 values of various models for the startup dataset - ')
result = {'Name of the ML Model':['Linear Regression',
                  'Ordinary Leaset Square (OLS) - raw',
                  'Standard transformed OLS',
                  'MinMax transformed OLS',
                  'Squareroot transformed OLS',
                  'OLS - improved'],
          'R^2 value':[r2_score(y_test, y_pred),
                      0.95,
                      0.95,
                      0.95,
                      0.932,
                      0.964]
         }
result_df = pd.DataFrame(result)
print(result_df)
```

Conclusion: 

Improved Ordinary Least Square model by removing outliers based on cook's threshold influence explains 96.4% of variation in dependent variable ie Profit of Startups.

#### Mult Linear Regression model

##### Linear regression

```python
# 1. Features selection
X = corolla_df.drop('Price', axis=1)
y = corolla_df['Price']
```

```python
# 2. Splitting of data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 3. Fitting ML model
# a) Linear regression 

car_regressor = LinearRegression()
car_regressor.fit(X_train, y_train)
y_preds = car_regressor.predict(X_test)
rmse_reg_car = (mean_squared_error(y_test, y_preds)) ** 0.5
print('R squared value of Regressor model: ', r2_score(y_test, y_preds))
print('Root Mean Square Error of Regressor model: ', rmse_reg_car)
```

##### Ordinary Least Square

```python
# b) Ordinary Least Square model

ox = sm.add_constant(X)

# Splitting data
ox_train, ox_test, oy_train, oy_test = train_test_split(ox, y, test_size=0.2, random_state=0)

ols_model = sm.OLS(oy_train, ox_train).fit()

print(ols_model.summary2())
```

Observations: 
1. Among Linear regression and Ordinary Least Squares (OLS) models, the OLS has higher R-square value of 0.866 compared to Linear regression model R-square value of 0.498. This shows that the OLS model explains the variation in dependent variable better than Linear regressoion.

2. The p-values for the t-test of all the features except Doors are below 0.05. This indicates that there is statistically significant relationship (at significance value of 0.05) between the features and Price (dependent variable).  

3. The probability value of F-statistic of the model is 0.00 (<0.05) which indicates that the model is statistically significant.

#### Observation of collinearity

```python
# Calculating VIF
vif = pd.DataFrame()
vif["Variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print('Variance Inflation Factors - ')
vif
```

Observation: 

VIF is more than 10 for all independent variables except KM. This shows that multicollinearity exists in the dataset.

#### OLS with data transformations

##### MinMax scaler

```python
# b) MinMax transformation
# Data transformation
mm_scaler = MinMaxScaler()
X_train_mm = mm_scaler.fit_transform(X)
X_train_mm = sm.add_constant(X_train_mm)
Y_train_mm = mm_scaler.fit_transform(pd.DataFrame(y))

# Fit the regression model
mm_model = sm.OLS(Y_train_mm, X_train_mm).fit()
print(mm_model.summary2())
```

##### Outliers treated

```python
# Improved data feature extraction
x_new2 = new_data2.drop(['Price', 'Influence'], axis=1)
y_new2 = new_data2['Price']

# Standard data transformation
x_new2_scaled = std_scaler.fit_transform(x_new2)
y_new2_scaled = std_scaler.fit_transform(pd.DataFrame(y_new2))

#Adding constant
x_new2_scaled = sm.add_constant(x_new2_scaled)

# Model fitting
new_data2_model = sm.OLS(y_new2_scaled, x_new2_scaled).fit()
print(new_data2_model.summary2())
```
